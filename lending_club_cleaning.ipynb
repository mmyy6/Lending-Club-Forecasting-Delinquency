```python
# Exporting loan_trunc to csv version as a record
loan_trunc.to_csv('loan_trunc.csv', index=False)
```

```python
# Using info() to see the number of non-nulls and the type for each column
loan_trunc.info(verbose=True, show_counts=True)
```

```python
# Remove the rows that have nulls which prevent effective analysis of data

loan_trunc = loan_trunc.dropna(subset=['emp_title'])
loan_trunc = loan_trunc.dropna(subset=['emp_length'])
loan_trunc = loan_trunc.dropna(subset=['percent_bc_gt_75'])
loan_trunc = loan_trunc.dropna(subset=['num_accts_ever_120_pd'])
loan_trunc = loan_trunc.dropna(subset=['dti'])
loan_trunc = loan_trunc.dropna(subset=['revol_util'])
```

```python
# To double-check that all rows with nulls have been removed (with some exceptions).
# The remaining rows with nulls are due to valid reasons (e.g. the application is not a joint application, hence there is no joint debt to income ratio, etc.)

loan_trunc.info(verbose=True, show_counts=True)
```

```python
loan_trunc.home_ownership.unique()

# The home ownership column has 2 responses (i.e. 'ANY', 'NONE') that are not the standard response options given by the bank.
```

```python
# To clean up the 2 responses, we have decided to replace them with "OTHER"

loan_trunc.replace('ANY', 'OTHER', inplace=True)
loan_trunc.replace('NONE', 'OTHER', inplace=True)
```

```python
loan_trunc.home_ownership.unique()
```

```python
# Checking for the number of nulls within the 'mths_since_last_delinq' column

loan_trunc.mths_since_last_delinq.isnull().sum()
```

```python
# Imputing the nulls with 999 as the nulls mean that the borrower has never been delinquent

loan_trunc['mths_since_last_delinq'] = loan_trunc['mths_since_last_delinq'].fillna(999)
```

```python
loan_trunc['mths_since_last_delinq'].unique()
```

**Investigating outliers**

```python
# int_rate

loan_trunc[loan_trunc.int_rate > 25]

# When filtered to interest rates that are above 25%, there were 44,332 loans. Therefore, the max of 30.99 does not seem to be an error.
```

```python
# dti

loan_trunc[loan_trunc.dti > 100].dti.describe()

# The debt to income column is unusually high. A dti of 100 would mean that for every $100 of monthly income, the borrower has $100 in debt. As such, it is unlikely that the dti above 100 is accurate.
# The rows with dti above 100 should be removed as they cannot be used in the predictive model.

loan_trunc = loan_trunc[loan_trunc.dti < 100]
```

```python
# delinq_2yrs

loan_trunc[loan_trunc.delinq_2yrs > 5]

# When filtered to more than 5 delinquencies in the past 2 years, there were more than 8000 borrowers. As such, this is not an error and no further action is required.
```

```python
# mort_acc

loan_trunc[loan_trunc.mort_acc > 90]

# There is 1 borrower with a total of 94 mortgage accounts. While it might seem like an error that a borrower has more than 90 mortgage accounts, it is noted that there are a total of 4 borrowers with more than 50
# mortgages, and 22 borrowers with more than 30 mortgage accounts. As such, we assume that this was not an error and have decided to keep this in the dataset.
```

```python
# num_accts_ever_120_pd

loan_trunc[loan_trunc.num_accts_ever_120_pd > 50]

# This does not appear to be an error. The borrowers with more than 50 defaulted accounts have the purpose of 'debt consolidation' - which means they likely have a history of having numerous loans in the past.
```

```python
# pub_rec

loan_trunc[loan_trunc.pub_rec > 50]

# The outlier of 86 public derogatory records for one person seems to be an error. However, when lowering the filter to 50 records, there appears to be other borrowers with similar records.
# As such, no change is made here.
```

```python
# annual_inc

loan_trunc[loan_trunc.annual_inc > 1000000]

# There seems to be an error with the annual income column as the high annual incomes do not correspond with the employment title of the borrower.
# For example, there is a social worker with an annual income of $9 million.
```

```python
# As there is no reliable way of verifying the data or fixing it, we have opted to remove all borrowers with income above $1 million.

loan_trunc = loan_trunc[loan_trunc.annual_inc < 1000000]
```

```python
# Exporting loan_clean to csv version as a record
loan_clean = loan_trunc
loan_clean.to_csv('loan_clean.csv', index=False)
```
